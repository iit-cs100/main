```python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv('survey.tsv', sep='\t')
print('%d responses' % len(df))
display(df)
display(df.describe())
```

```python
for question in df.columns:
    plt.figure()
    plt.hist(df[question], bins=np.arange(7)-.5, width=1, edgecolor='black')
    plt.xlim(.5,5.5)
    #plt.xticks(np.arange(6)+1)
    plt.title(question, size=14)
    plt.ylabel('# students', size=14)
    plt.xlabel('rating (5=agree)', size=14)
    plt.show()
```
